```python
import matplotlib.pyplot as plt
# 별칭 = as 
```

```python
x = [1,2,3,4]
y = [2,3,4,5]

# 그림을 그리기 위하여 차트를 먼저 정의
# plot() : 꺾은선 그래프
# plt.show() = print()의 개념과 같이 보여줌.

plt.plot(x, y)
plt.show()
```

```python
# bar() : 막대그래프
plt.bar(x,y)
plt.show()
```

```python
# 차트의 위치 처리
# figure() : 그리는영역 정의, 데이터 분할저장
figure = plt.figure()
axes1 = figure.add_subplot(2,2,1) # 1번위치
axes2 = figure.add_subplot(2,2,2) # 2번위치
axes3 = figure.add_subplot(2,2,3) # 3번위치
axes4 = figure.add_subplot(2,2,4) # 4번위치
```

```python
import numpy as np
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,4,6,8]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,7])

# axes.plot(x,y)
# axes.plot(x2,y2)

# color : 그래프선 색상 변경 
# linestyle : 그래프선 스타일 변경 (dotted:점선,dashed:끊킨선(?)
# marker : 해당지점 표시 
axes.plot(x,y, color='red', linestyle='dotted', marker='o')
axes.plot(x2,y2, color='green', linestyle='dashed', marker='^')
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,3,5,7]
y = [2,4,6,8]

x2 = [2,4,6,8]
y2 = [4,4,3,6]
#같은위치에 있으면 중첩되어 2번째 그래프가 1번째 그래프를 가린다. 

axes.bar(x,y, color='violet')
axes.bar(x2,y2, color='gray')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]


axes.bar(x,y, color='lightcoral', label='bar')
axes2.plot(x2,y2, color='blue', label='plot')

# legend() : label값 출력
axes.legend()
axes2.legend(loc=3)
# loc=1(우측상단) 2(좌측상단) 3(좌측하단) 4(우측하단)

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,4,6,8]

x2 = [1,2,3,4]
y2 = [4,6,1,9]


# scatter() : 분포도 만들때 사용! 
axes.scatter(x,y, color='lightcoral')
axes.scatter(x2,y2, color='green')


plt.show()
```

```python
from matplotlib import font_manager, rc

font_path = ''
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

figure = plt.figure()
axes = figure.add_subplot(111)

# 한글 사용시 깨져서 나옴 위의 matplotlib 깔아주기
label = ['자바','파이썬','자바스크립트','리액트']
data = [1,2,5,2]

# pie() : 원형분포도 표 만들기 
axes.pie(data, labels=label)

# plt.show()

# 파일 저장시 show() 명령어를 주석처리 해줘야 정상적으로 저장이 된다. 아니면 백지로 저장됨.
plt.savefig('test')
```
